```python
import torch
import torchvision
```

```python
cifar = torchvision.datasets.CIFAR10("./data/cifar10/", train=True, download=True)
print(type(cifar.data))
print(type(cifar.data.shape))
```

```python
x = torch.from_numpy(cifar.data)
x = x.transpose(1, 3)
x = x.transpose(2, 3)
x = x.narrow(0, 0, 48)
x = x.float()
x = x.div(255)
x = x.transpose(2, 3)
# x.narrow(1, 1, 2).fill_(0)
x[:,1:3,:].fill_(0)
print()
```

```python
from IPython.display import Image
torchvision.utils.save_image(x, 'file_name.png', nrow=12)
Image(filename='file_name.png')
```
